```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
```

## Session Spark :

```python
spark = SparkSession.builder \
    .appName("BDA - Price Data Ingestion") \
    .master("local[*]") \
    .getOrCreate()
```

## Read data :

```python
price_data_path = "../data/prices/btcusd_1-min_data.csv"
df_prices = spark.read.csv(price_data_path, header=True, inferSchema=True)
print(f"Showing first 5 rows : {price_data_path}")
df_prices.show(5)
```

```python
print("Schematic df :")
df_prices.printSchema()
```

```python
row_count = df_prices.count()
print(f"\n Total of {row_count:,} rows.")
```
